```python
#Importing essential libraries including matplotlib for visualization,
#easyocr for optical charcter recognition, 
#cv2 for reading and displaying images,
#and rcParams is used to change the size of figures.

import matplotlib.pyplot as plt
import easyocr
import cv2
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize']=8,16
```

```python
#Downloading the model for English language
reader=easyocr.Reader(['en'])
```

```python
Image("Picture2.jpg")
```

```python
#Using readtext function to extract text from our input image
output=reader.readtext('Picture2.jpg')
```

```python
output
```

```python
#Here we see the bounding box coordinates; text generated; and confidence respectively.
#Lets try and see where the bounding box was created.
```

```python
box = output[-1][0]
```

```python
x_min,y_min=[min(idx) for idx in zip(*box)]
```

```python
x_max,y_max=[max(idx) for idx in zip(*box)]
```

```python
image=cv2.imread('Picture2.jpg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
